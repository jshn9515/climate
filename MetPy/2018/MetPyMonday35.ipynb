## 35 - Matplotlib Transforms

Watch video: [YouTube](https://www.youtube.com/watch?v=CxsShxafkM8)

Basically, Matplotlib has two transforms: `ax.transAxes` and `fig.transFigure`.
- `ax.transAxes`: This attribute is associated with an Axes object (ax) and represents the coordinate transformation relative to the Axes' bounding box. In other words, it maps the coordinates within the range \[0, 1] where (0, 0) is the bottom-left corner and (1, 1) is the top-right corner of the Axes.
- `fig.transFigure`: This attribute belongs to a Figure object (fig) and defines the coordinate transformation relative to the Figure's bounding box. Similar to ax.transAxes, it maps coordinates within the range \[0, 1], where (0, 0) is the bottom-left corner and (1, 1) is the top-right corner of the Figure.

```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = range(1, 11)
y = [math.log(i) for i in x]
```

```python
plt.rcParams['font.sans-serif'] = ['Arial']
fig = plt.figure(1, figsize=(8, 6))
ax = plt.subplot(1, 1, 1)
ax.plot(x, y, linestyle='None', marker='o')
ax.text(0.5, 0.95, 'Top Center Label - My Plot', fontsize=16, va='center', ha='center', transform=ax.transAxes)
# You can change the `xlim` whatever you like, but the text will always remain at the top center of the axes.
ax.set_xlim(0, 10)
plt.show()
```

```python
fig = plt.figure(2, figsize=(8, 6))
ax = plt.subplot(1, 1, 1)
ax.plot(x, y, linestyle='None', marker='o')
ax.text(0.5, 0.92, 'Top Center Label - My Plot', fontsize=16, va='center', ha='center', transform=fig.transFigure)
# You can change the `xlim` whatever you like, but the text will always remain at the top center of the figure.
ax.set_xlim(0, 10)
plt.show()
```
