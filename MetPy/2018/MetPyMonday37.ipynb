## 37 - Twin x-Axes

Watch video: [YouTube](https://www.youtube.com/watch?v=lrREMkpHfao)

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.arange(100)
y1 = np.random.random(100)
y2 = np.random.random(100) * 100
```

```python
plt.figure(1)
plt.plot(x, y1)
plt.show()
```

```python
plt.figure(2)
plt.plot(x, y2)
plt.show()
```

```python
fig = plt.figure(3, figsize=(8, 6))
ax1 = fig.add_subplot(1, 1, 1)
ax2 = ax1.twinx()
ax1.plot(x, y1, color='tab:blue')
ax2.plot(x, y2, color='tab:orange')

ax1.set_ylabel('variable 1', fontsize=16)
ax1.yaxis.label.set_color('tab:blue')
ax1.tick_params(axis='y', colors='tab:blue')
ax2.spines['left'].set_color('tab:blue')

ax2.set_ylabel('variable 2', fontsize=16, rotation=-90, labelpad=13)
ax2.yaxis.label.set_color('tab:orange')
ax2.tick_params(axis='y', colors='tab:orange')
ax2.spines['right'].set_color('tab:orange')
plt.show()
```
