````python
import numpy as np
import matplotlib.pyplot as plt
from metpy.plots import ctables
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
````

````python
x, y = np.mgrid[-5: 5: 0.05, -5: 5: 0.05]
print(x)
````

````python
z = 5 * (np.hypot(x, y) + np.sin(x ** 2 + y ** 2))
````

````python
plt.figure(1)
plt.plot(x[:, 100], z[:, 100])
plt.show()
````

````python
np.max(z)
````

````python
plt.figure(2)
plt.imshow(z)
plt.show()
````

To get the list of color table, type the following command:
```python
print(ctables.registry.keys())
```

````python
plt.figure(3)
norm, cmap = ctables.registry.get_with_range('Carbone42', 0, 40)
plt.imshow(z, norm=norm, cmap=cmap)
plt.show()
````
