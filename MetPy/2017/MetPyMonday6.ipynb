## 6 - Making a Basic Map with Cartopy

Watch video: [YouTube](https://www.youtube.com/watch?v=iPJjtU7OFSQ)

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
fig = plt.figure(1, figsize=(8, 10))
ax = fig.add_subplot(projection=ccrs.LambertConformal())
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE, linewidth=0.8)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAKES, alpha=0.5)
ax.add_feature(cfeature.RIVERS)
ax.add_feature(cfeature.STATES.with_scale('50m'), linestyle='dotted', linewidth=0.5)
ax.set_extent([-120, -70, 20, 50])
plt.show()
```

```python
proj = ccrs.Stereographic(central_longitude=-95, central_latitude=35)
plt.figure(2)
ax = plt.subplot(projection=proj)
ax.add_feature(cfeature.STATES.with_scale('50m'), facecolor='none', edgecolor='black')
ax.add_feature(cfeature.BORDERS)
ax.gridlines()
ax.set_extent([-109, -100, 35, 42])
ax.scatter(-105.27, 40.01, transform=ccrs.PlateCarree(), marker='*', s=400, c='red')
plt.show()
```
